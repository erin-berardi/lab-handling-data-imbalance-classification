```python
import pandas as pd
import numpy as np
import warnings
import statsmodels.api as sm
import sklearn
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression

pd.set_option('display.max_columns', None)
warnings.filterwarnings('ignore')
```

```python
donors = pd.read_csv('numerical.csv')
categorical = pd.read_csv('categorical.csv')
numerical = pd.read_csv('numerical.csv')
```

```python
donors = pd.concat([donors, pd.read_csv('categorical.csv')], axis=1)
```

```python
categorical = categorical.dropna()
```

```python
scaler = StandardScaler()
numerical_scaled = scaler.fit_transform(numerical) 
scaled_numerical = pd.DataFrame(numerical_scaled, columns=numerical.columns)
```

```python
scaled_numerical
```

```python
categorical = pd.get_dummies(categorical)
```

```python
categorical
```

```python
X = pd.concat([scaled_numerical, categorical], axis=1)
y = pd.read_csv('Y.csv')
y = y.drop('TARGET_D', axis=1)
```

```python
X
```

```python
null_indexes = X[X.isnull().any(axis=1)].index

print(null_indexes)
```

```python
X = X.dropna()
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
```

```python
model = LogisticRegression()

model.fit(X_train, y_train)

y_pred = model.predict(X_test)

accuracy = model.score(X_test, y_test)

print("Accuracy:", accuracy)
```

```python
y.value_counts()
```

```python
from imblearn.under_sampling import RandomUnderSampler

rus = RandomUnderSampler(random_state=0)
X_under_resampled, y_under_resampled = rus.fit_resample(X, y)
X_under_train,X_under_test,y_under_train,y_under_test = train_test_split(X_under_resampled,y_under_resampled,test_size=0.2)
```

```python
y_under_resampled.value_counts()
```

```python
under_model = LogisticRegression()

under_model.fit(X_under_train, y_under_train)

under_y_pred = under_model.predict(X_under_test)

accuracy = under_model.score(X_under_test, y_under_test)

print("Accuracy:", accuracy)
```

```python
from imblearn.over_sampling import SMOTE

smote = SMOTE()

X_over_resampled, y_over_resampled = smote.fit_resample(X, y)
X_over_train,X_over_test,y_over_train,y_over_test = train_test_split(X_over_resampled,y_over_resampled,test_size=0.2)
y_over_resampled.value_counts()
```

```python
over_model = LogisticRegression()

over_model.fit(X_over_train, y_over_train)

over_y_pred = over_model.predict(X_over_test)

accuracy = over_model.score(X_over_test, y_over_test)

print("Accuracy:", accuracy)
```

```python
from imblearn.under_sampling import TomekLinks
tl = TomekLinks(sampling_strategy='majority')
X_tl, y_tl = tl.fit_resample(X, y)
y_tl.value_counts()

X_tl_train,X_tl_test,y_tl_train,y_tl_test = train_test_split(X_tl,y_tl,test_size=0.2)
```

```python
tl_model = LogisticRegression()

tl_model.fit(X_tl_train, y_tl_train)

tl_y_pred = tl_model.predict(X_tl_test)

accuracy = tl_model.score(X_tl_test, y_tl_test)

print("Accuracy:", accuracy)
```
