```python
# 1. Import the required libraries and modules that you would need.
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import OneHotEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from imblearn.under_sampling import RandomUnderSampler
from imblearn.over_sampling import SMOTE

import warnings
warnings.filterwarnings("ignore")
```

```python
# 2. Read that data into Python and call the dataframe donors.
# 3. Check the datatypes of all the columns in the data.
# 4. Check for null values in the dataframe. Replace the null values using the methods learned in class.
# 5. Split the data into numerical and catagorical. Decide if any columns need their dtype changed.
# 6. Concatenate numerical and categorical back together again for your X dataframe. Designate the Target as y.
numerical = pd.read_csv('learningSet_numerical_clean.csv')
categorical = pd.read_csv('learningSet_categorical_clean.csv')
target = pd.read_csv('learningSet_Y.csv')
```

```python
    # 6.1. Split the data into a training set and a test set.
    # 6.2. Split further into train_num and train_cat. Also test_num and test_cat.
    # 6.3. Scale the features either by using normalizer or a standard scaler. (train_num, test_num)Encode the categorical features using One-Hot Encoding or Ordinal Encoding. (train_cat, test_cat)
    # 6.4. Encode the categorical features using One-Hot Encoding or Ordinal Encoding. (train_cat, test_cat)
        # - fit only on train data transform both train and test
        # - again re-concatenate train_num and train_cat as X_train as well as test_num and test_cat as X_test

# Above order of works does not makes sense. Will split into train/test after scalling and OneHotEncoding
```

```python
# Scalling
scaler = StandardScaler()
scaler.fit(numerical)
numerical_prep = pd.DataFrame(scaler.transform(numerical))
numerical_prep.columns = numerical.columns
numerical_prep
```

```python
# One Hot Encoding
categorical.nunique()
```

```python
categorical = categorical.drop(columns='ZIP')
```

```python
onehot_cols = []
for col in categorical.columns:
    col_uniques = sorted(categorical[col].astype(str).unique())
    for unique in col_uniques:
        onehot_cols.append(col+"_"+unique)

encoder = OneHotEncoder().fit(categorical)
encoded_categorical = pd.DataFrame(encoder.transform(categorical).toarray())
encoded_categorical.columns = onehot_cols
encoded_categorical
```

```python
# Concatenating and train-test split
X = pd.concat((numerical_prep,encoded_categorical),axis=1).reset_index(drop=True)
X
```

```python
target=target.drop(columns="TARGET_D")
X_train, X_test, y_train, y_test = train_test_split(X,target, test_size=0.2, random_state=100)
```

```python
    # 6.5. Fit a logistic regression model on the training data.
model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
    # 6.6. Check the accuracy on the test data.
model.score(X_test, y_test)
```

### Upsampling

```python
data = pd.concat((target,X),axis=1).reset_index(drop=True)
data
```

```python
category_0 = data[data['TARGET_B']==0]
category_1 = data[data['TARGET_B']==1]

category_1_new = category_1.sample(round(len(category_0)*0.8),replace=True)
data2 = pd.concat([category_0,category_1_new],axis=0)
data2['TARGET_B'].value_counts()
```

```python
X = data2.drop(columns='TARGET_B')
y = data2['TARGET_B']

X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2, random_state=100)
model = LogisticRegression()
model.fit(X_train, y_train)
model.score(X_test, y_test)
```

### Downsampling

```python
category_0 = data[data['TARGET_B']==0]
category_1 = data[data['TARGET_B']==1]

category_0_new = category_0.sample(len(category_1)*2)

data2 = pd.concat([category_0_new,category_1],axis=0)
data2['TARGET_B'].value_counts()
```

```python
X = data2.drop(columns='TARGET_B')
y = data2['TARGET_B']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2, random_state=100)
model = LogisticRegression()
model.fit(X_train, y_train)
model.score(X_test, y_test)
```

```python
# Downsampling at len(a) = len(2b) yields a slightly better score than upsampling at len(b) = len(a*0.8)
```
